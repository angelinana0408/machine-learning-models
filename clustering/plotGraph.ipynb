```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict



def getnpdatasetFromCSV(filePath):
    dataframe = pd.read_csv(filePath)
    npdataSet = dataframe.values
    return npdataSet

def getnpdatasetFromTXT(filePath):
        dataSet = []
        with open(filePath) as f:
            lines = f.readlines()
        lines = [x.strip() for x in lines]
        for line in lines:
            points = line.split('\t')
            dataSet.append(points)
        npdataset = np.array(dataSet)
        return npdataset

def plotGraph(dataset, graphTitle):
    style = ['bo', 'go', 'ro', 'co', 'mo', 'yo', 'ko','bx', 'gx', 'rx', 'cx', 'mx', 'yx', 'kx']
    classID = set()
    #get classes
    for point in dataset:
        classID.add(point[2])
    #plot points
    styleIndex = 0
    for classNo in classID:
        for point in dataset:
            if point[2] == classNo:    
                plt.plot(point[0],point[1],style[styleIndex], label=graphTitle+" "+str(int(classNo)) )                
        styleIndex += 1
    plt.xlabel('x1 feature')
    plt.ylabel('x2 feature')
    
    plt.title(graphTitle)
    
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    keylist = by_label.keys()
    by_label_sorted = {}
    for key in sorted(keylist):
        by_label_sorted.update({key: by_label[key]})
    plt.legend(by_label_sorted.values(), by_label_sorted.keys(), bbox_to_anchor=(1.05, 1))
    
    plt.show()
    

    
print("===========================DataSet 1===========================")
#ground truth data
labelNpDataSet = getnpdatasetFromTXT("dataset1.txt")
plotGraph(labelNpDataSet, "ground truth label")
ClusterNpDataSet = getnpdatasetFromCSV("Kmeans_1.csv")
plotGraph(ClusterNpDataSet, "KMeans calculated cluster")
ClusterNpDataSet = getnpdatasetFromCSV("DBSCAN_dataset1.csv")
plotGraph(ClusterNpDataSet, "DBSCAN calculated cluster")
ClusterNpDataSet = getnpdatasetFromCSV("GMM1.csv")
plotGraph(ClusterNpDataSet, "GMM calculated cluster")

print("===========================DataSet 2===========================")
#ground truth data
labelNpDataSet = getnpdatasetFromTXT("dataset2.txt")
plotGraph(labelNpDataSet, "ground truth label")
ClusterNpDataSet = getnpdatasetFromCSV("Kmeans_2.csv")
plotGraph(ClusterNpDataSet, "KMeans calculated cluster")
ClusterNpDataSet = getnpdatasetFromCSV("DBSCAN_dataset2.csv")
plotGraph(ClusterNpDataSet, "DBSCAN calculated cluster")
ClusterNpDataSet = getnpdatasetFromCSV("GMM2.csv")
plotGraph(ClusterNpDataSet, "GMM calculated cluster")

print("===========================DataSet 3===========================")
#ground truth data
labelNpDataSet = getnpdatasetFromTXT("dataset3.txt")
plotGraph(labelNpDataSet, "ground truth label")
ClusterNpDataSet = getnpdatasetFromCSV("Kmeans_3.csv")
plotGraph(ClusterNpDataSet, "KMeans calculated cluster")
ClusterNpDataSet = getnpdatasetFromCSV("DBSCAN_dataset3.csv")
plotGraph(ClusterNpDataSet, "DBSCAN calculated cluster")
ClusterNpDataSet = getnpdatasetFromCSV("GMM3.csv")
plotGraph(ClusterNpDataSet, "GMM calculated cluster")
```
