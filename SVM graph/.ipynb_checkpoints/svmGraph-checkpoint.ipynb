```python
import sys
import numpy as np
import matplotlib.pyplot as plt

def load_file(file_name):
    dataset = []
    with open(file_name, 'r') as f:
        for line in f:
            data = line.strip().split(' ')
            for index in range(len(data)):
                data[index] = float(data[index])
            dataset.append(data)
    return dataset
def f(x1):
    result = (2.68-1.8758*x1)/0.8147
    return result

def plotGraph(dataset):
    #plot points
    for index in range(len(dataset)):
        if dataset[index][2] == 1:
            positive, = plt.plot(dataset[index][0], dataset[index][1], 'ro', label='positive')
        else:
            negative, = plt.plot(dataset[index][0], dataset[index][1], 'go', label='negative')
    plt.xlabel('x1 feature')
    plt.ylabel('x2 feature')
    plt.legend([positive, negative], ['positive', 'negative'])
    
    #plot boundary
    x1 = np.arange(-2.0, 5.0, 0.1)
    plt.plot(x1, f(x1))
    
    plt.show()
    
filename = 'svmdata.txt'
dataset = load_file(filename)
plotGraph(dataset)
```
